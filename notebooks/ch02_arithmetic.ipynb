## Try it yourself

Change the number of players below and see how the teams and leftovers change.

```python
total_points = 2 + 3 * 4
teams = 17 // 5
leftover_players = 17 % 5
print(total_points, teams, leftover_players)
```

## Now make it your own

```python
def split_into_teams(players, team_size):
    return players // team_size, players % team_size


print(split_into_teams(25, 4))
```
