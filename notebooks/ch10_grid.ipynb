The coordinate plane from math class – an x and a y telling you exactly where a point is – is exactly how a 2D game world tracks where everything stands.

**By the end of this chapter you can:**
- Represent a position as an (x, y) pair
- Update a position by adding a movement delta
- Explain how a 2D grid maps to the game screen

## Try it yourself

Change dx and dy below to move the point in different directions.

```python
player_pos = (2, 3)
dx, dy = 1, 0
new_pos = (player_pos[0] + dx, player_pos[1] + dy)
print(new_pos)
```

## Now make it your own

```python
def move(pos, dx, dy):
    return (pos[0] + dx, pos[1] + dy)


print(move(player_pos, 0, 1))
```
