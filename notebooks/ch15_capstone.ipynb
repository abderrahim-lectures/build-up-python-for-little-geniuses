## Try it yourself

This is the last notebook in the book – change hero's name and inventory to make it your own.

```python
class Player:
    def __init__(self, name, health):
        self.name = name
        self.health = health


hero = Player("Ari", 100)
inventory = ["sword"]
prices = {"potion": 5}

for item in inventory:
    print(f"{hero.name} is carrying a {item}")
print(f"A potion costs {prices['potion']} coins")
```

## Now make it your own

```python
def build_summary(name, inventory, prices):
    items = ", ".join(inventory)
    cheapest = min(prices.values()) if prices else 0
    return f"{name} is carrying: {items}. Cheapest shop item: {cheapest} coins."


print(build_summary(hero.name, inventory, prices))
```
