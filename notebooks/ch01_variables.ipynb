A variable is like a labelled storage crate: put a value inside, write a name on it, and Python remembers it until you change it.

**By the end of this chapter you can:**
- Store a value in a variable
- Explain why we give variables meaningful names
- Update a variable and see the new value take effect

## Try it yourself

Change `player_name` below to your own name, then run both cells.

```python
player_name = "Ari"
score = 0
health = 100
print(player_name, score, health)
```

## Now make it your own

```python
def describe_player(name, score, health):
    return f"{name} - score: {score}, health: {health}"


print(describe_player(player_name, score, health))
```
