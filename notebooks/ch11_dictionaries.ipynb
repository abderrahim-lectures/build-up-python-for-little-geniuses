A lookup table – like a times table where you look up 4 × 7 and get 28 – pairs a key with a value; Python's dict does exactly that.

**By the end of this chapter you can:**
- Create a dictionary and look up a value by key
- Add and update key-value pairs
- Explain when a dict is better than a list

## Try it yourself

Add a new item and price to the dictionary below.

```python
prices = {"potion": 5, "sword": 20, "shield": 15}
prices["potion"] = 4
print(prices["sword"])
print(prices)
```

## Now make it your own

```python
def total_cost(cart, prices):
    return sum(prices[item] for item in cart)


print(total_cost(["sword", "shield"], prices))
```
