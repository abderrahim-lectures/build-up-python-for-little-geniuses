A running total that keeps going until you reach a target is exactly what a while loop does: repeat until a condition becomes false.

**By the end of this chapter you can:**

## Try it yourself

Change the starting health below and watch the countdown.

```python
health = 5
while health > 0:
    print(health)
    health = health - 1
print("game over")
```

## Now make it your own

```python
def countdown(start):
    values = []
    health = start
    while health > 0:
        values.append(health)
        health -= 1
    return values


print(countdown(7))
```
