Grouping related facts about one thing – a shape's length AND width together – is what a class does: it bundles data and behavior into one blueprint.

**By the end of this chapter you can:**
- Define a class with __init__ and a method
- Create an object (instance) from a class
- Explain the difference between a class and an object

## Try it yourself

Create your own Player below and take some damage.

```python
class Player:
    def __init__(self, name, health):
        self.name = name
        self.health = health

    def take_damage(self, amount):
        self.health = max(0, self.health - amount)


hero = Player("Ari", 100)
hero.take_damage(30)
print(hero.name, hero.health)
```
