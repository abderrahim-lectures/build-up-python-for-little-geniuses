Word problems mix numbers and words – Python's f-strings let your code do the same, dropping a variable's value right into a sentence.

**By the end of this chapter you can:**

## Try it yourself

Change the name and score below, then run both cells.

```python
player_name = "Ari"
score = 10
message = f"{player_name} has {score} points"
print(message)
print(len(player_name))
```

## Now make it your own

```python
def player_status(name, score):
    return f"{name} has {score} points"


print(player_status(player_name, score))
```
