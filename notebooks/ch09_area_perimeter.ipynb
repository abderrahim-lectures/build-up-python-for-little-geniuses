Fencing off a rectangular plot of land needs its perimeter; covering it with grass needs its area – both are just formulas turned into functions.

**By the end of this chapter you can:**

## Try it yourself

Change plot_length and plot_width below and compare area vs. perimeter.

```python
def area(length, width):
    return length * width

def perimeter(length, width):
    return 2 * (length + width)


plot_length, plot_width = 6, 4
print("area:", area(plot_length, plot_width))
print("perimeter:", perimeter(plot_length, plot_width))
```
