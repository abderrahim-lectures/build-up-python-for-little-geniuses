A number line splits into 'less than', 'equal to', and 'greater than' – if/elif/else lets your code make exactly that kind of choice.

**By the end of this chapter you can:**
- Write an if/elif/else chain
- Use comparison operators (<, >, ==)
- Explain why only one branch of an if/elif/else ever runs

## Try it yourself

Change `health` below and see which branch runs.

```python
health = 15
if health <= 0:
    status = "defeated"
elif health < 25:
    status = "low health"
else:
    status = "ok"
print(status)
```

## Now make it your own

```python
def health_status(health):
    if health <= 0:
        return "defeated"
    elif health < 25:
        return "low health"
    else:
        return "ok"


print(health_status(25))
```
