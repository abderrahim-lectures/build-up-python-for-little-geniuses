Rolling a die or flipping a coin is probability in action – Python's random module lets your code do the same roll.

**By the end of this chapter you can:**

## Try it yourself

Run the cell below a few times – the numbers should change each time.

```python
import random

roll = random.randint(1, 6)
loot = random.choice(["sword", "shield", "potion"])
print(roll, loot)
```

## Now make it your own

```python
def roll_die(sides=6):
    return random.randint(1, sides)


print(roll_die(20))
```
