Repeated addition – <code>4 + 4 + 4</code> – is multiplication; a for loop is Python's way of repeating an action a set number of times without writing it out by hand.

**By the end of this chapter you can:**
- Write a for loop over a range()
- Write a for loop over a list

## Try it yourself

Change how many times the loop repeats below.

```python
total = 0
for i in range(4):
    total = total + 3
print(total)
```

## Now make it your own

```python
def repeated_addition(value, times):
    total = 0
    for _ in range(times):
        total += value
    return total


print(repeated_addition(3, 6))
```
