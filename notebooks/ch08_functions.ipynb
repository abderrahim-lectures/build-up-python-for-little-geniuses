A math formula like <code>area = length * width</code> takes inputs and produces an output – a Python function is exactly that: input in, answer out.

**By the end of this chapter you can:**
- Define a function with def, parameters, and return
- Call a function with different arguments
- Explain why functions help avoid repeating code

## Try it yourself

Call area() with a few different lengths and widths below.

```python
def area(length, width):
    return length * width


print(area(4, 5))
```

## Now make it your own

```python
def move_player(x, dx):
    return x + dx


print(move_player(10, 3))
```
